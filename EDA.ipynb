```python
import seaborn as sns
sns.get_dataset_names()
```

```python
import matplotlib.pyplot as plt

#load planets dataset
p=sns.load_dataset('titanic')

#display first few rows
p.head(80)
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='age',data=p)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='age',y='sex',data=p)                        #x=axis -->categorical            y=axis-->int
plt.show()
```

```python
#kde=> kernel distribution estimation
plt.figure(figsize=(6,4))
sns.histplot(p['fare'],bins=20,kde=True,color='black')
plt.show()
```

```python
counts=p['survived'].value_counts()
counts
```

```python
#SMOT-SYNTHYETIC MINORITY OVERSAMOLING TECHNIQUE( imblanced data to balance data)
#i)oversampling             yes athigama iruntha
#ii)undersampling          no athigama iruntha

plt.figure(figsize=(4,6))
plt.pie(counts,labels=['Did Not survive','survived	'],colors=['skyblue','blue'],autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(4,6))
sns.boxplot(x='fare',data=p,palette='Set1')
plt.title('Fare Distribution by passengers class')                   #iqr

plt.xlabel('class')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(5,8))
sns.scatterplot(x='age',y='fare',hue='survived',style='sex',data=p,palette='rainbow')           #palette=coolwarm
plt.title('Age vs Fare with Survival and Gender')
plt.xlabel('age')
plt.ylabel('Fare')
plt.legend(title='Survived')
plt.show()
```

```python
numerical=p[['survived','pclass','age','sibsp','parch','fare']].corr()
numerical
```

```python
numerical=p[['survived','pclass','age','sibsp','parch','fare']].corr()

plt.figure(figsize=(4,4))
sns.heatmap(numerical,annot=True,cmap='coolwarm',center=1)                #color-->coolwarm=rainbow
plt.title('Correlation heatmap of Titanic Dataset')
plt.show()                               

#only int values
```

```python
plt.figure(figsize=(7,8))
sns.violinplot(x='class',y='age',hue='survived',data=p,split=False,palette='Set1')
plt.title('age Distribution by Class and Survival')
plt.xlabel('Class')
plt.ylabel('Age')
plt.legend(title='Survived')
plt.show()
#combination of box plot and KDE to show distribution shape ,spred and outliers(KDE-->KERNAL DENSITY ESTIMATION)
```

```python
sns.kdeplot(data=p['age'].dropna(),color='red',fill=True)
plt.title('Age Distribution')
plt.xlabel('fare')
plt.ylabel('density')
plt.show()
#only int
```

```python
sns.pairplot(p)
```

```python
sns.pairplot(p[['age','fare','class','survived','sex']])
```

```python
plt.figure(figsize=(5,5))
sns.swarmplot(x='class',y='age',hue='survived',data=p,palette='Set2',dodge=False)
plt.title('Age Distribution by class and survival (Swarm Plot') 
plt.xlabel('class')
plt.ylabel('age')
plt.legend(title='Survived')
plt.show()

#advanced model of violinplot
```

```python
g=sns.FacetGrid(p,col='class',row='survived',margin_titles=True)
g.map(sns.histplot,'age',bins=20, color='yellow')
g.fig.suptitle('age distribution by class and survival',y=1.03)
```
