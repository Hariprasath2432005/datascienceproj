```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()                                  #IMPORT LABELENCODER 
```

```python
df['gender']=le.fit_transform(df['gender'])
df['smoker']=le.fit_transform(df['smoker'])
df['region']=le.fit_transform(df['region'])
df['medical_history']=le.fit_transform(df['medical_history'])
df['family_medical_history']=le.fit_transform(df['family_medical_history'])
df['exercise_frequency']=le.fit_transform(df['exercise_frequency'])
df['occupation']=le.fit_transform(df['occupation'])
df['coverage_level']=le.fit_transform(df['coverage_level'])
                                                     #CONVERT THE DATA TYPE OBJECT TO INT BY USING LABELENCODER
```

```python
df.info()         
```

```python
df=df.rename(columns={'family_medical_history':'fmh'})
df=df.rename(columns={'exercise_frequency':'frequency'})
                                                        #RENAME THE COLUMNS 
```

```python
df.info()
```

```python
a=df.drop(['charges'],axis=1)
b=df['charges']
print(a)                                               #DROP THE OUTPUT
print(b)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()               
T=sc.fit_transform(a)                                       #NORMALIS THE DATASET BY USING STANDARDSCALER
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression          #IMPORT THE TRAIN AND TEST SPLIT
```

```python
a_train,a_test,b_train,b_test=train_test_split(a,b,test_size=0.2,random_state=25)
model=LinearRegression()
model.fit(a_train,b_train)                                #TRAIN THE LINEARREGREASSION MODEL WITHOUT PARAMETER
b_pred=model.predict(a_test)
print(b_pred)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(b_test,b_pred)                                #PREDICT THE R2 SCORE FOR LINEARREGRESSION TRAIN MODEL
print(r2)
```
