```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['id']=le.fit_transform(df['id'])
df['title']=le.fit_transform(df['title'])
df['type']=le.fit_transform(df['type'])
df['description']=le.fit_transform(df['description'])
df['age_certification']=le.fit_transform(df['age_certification'])
df['imdb_id']=le.fit_transform(df['imdb_id'])
```

```python
a=df.drop(['imdb_score'],axis=1)
b=df['imdb_score']
print(a)
print(b)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
T=sc.fit_transform(a)
print(T)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
a_train,a_test,b_train,b_test=train_test_split(a,b,test_size=0.2,random_state=25)
model=LinearRegression()
model.fit(a_train,b_train)
b_pred=model.predict(a_test)
print(b_pred)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(b_test,b_pred)
print(r2)
```
