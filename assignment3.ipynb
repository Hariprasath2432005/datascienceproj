```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
df['state']=oe.fit_transform(df[['state']])
df['state_po']=oe.fit_transform(df[['state_po']])
df['office']=oe.fit_transform(df[['office']])
df['party_simplified']=oe.fit_transform(df[['party_simplified']])   
df.info()
```

```python
a=df.drop(['state_po'],axis=1)
b=df['state_po']
print(a)
print(b)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()                #this is used for normalize the dataset
T=sc.fit_transform(a)
print(T)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
a_train,a_test,b_train,b_test=train_test_split(a,b,test_size=0.2,random_state=25)
model=LinearRegression()
model.fit(a_train,b_train)
b_pred=model.predict(a_test)
print(b_pred)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(b_test,b_pred)
print(r2)
```
